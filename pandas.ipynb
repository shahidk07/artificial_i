# Working with timeseries data

```python
names=['Shahid','Kabir','David','Ishmael']
```

```python
dates=pd.date_range(start='12-01-2025',end='12-25-2025').astype(str)
```

```python
timedata=pd.DataFrame({
'Date':np.random.choice(dates,size=7),
    'Customer Name':np.random.choice(names,size=7),
    'Lemonades':np.random.randint(1,10,size=7)
})
```

```python
timedata
timedata['Date'].info()
```

```python
timedata
```

```python
timedata['Date']=pd.to_datetime(timedata['Date'])
```

```python
timedata.info()
```

```python
pd.to_datetime('2025-12-25')
```

```python
timedata
```

# Resampling

```python
timedata.resample('YE',on='Date').agg({'Lemonades':'sum',
                                      'Customer Name':lambda x:','.join(x.unique())})
```
