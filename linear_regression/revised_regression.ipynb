```python
import math
import numpy as np
import matplotlib as plt
```

```python
# Load our data set
x_train = np.array([1.0, 2.0])   #features
y_train = np.array([300.0, 500.0])   #target value
```

```python
def computecost(x,y,w,b):
    m=x.shape[0]
    cost=0
    for i in range(m):
        f_wb=w*x[i]+b
        cost+=(f_wb-y[i])**2
    total_cost=cost/(2*m)
    return total_cost 
```

```python
def computegradient(x,y,w,b):
    m=x.shape[0]
    dj_db=0
    dj_dw=0
    for i in range(m):
        f_wb=w*x[i]+b
        dj_dw+=(f_wb-y[i])*x[i]
        dj_db+=(f_wb-y[i])
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_dw,dj_db
    
    
```

```python
def gradientdescent(x,y,w,b,alpha,iterations,computecost,computegradient):

    m=x.shape[0]
    p_history=[]
    j_history=[]
    for i in range(iterations):
        dj_dw,dj_db=computegradient(x,y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
        
        if(i<iterations):
            p_history.append([w,b])
            j_history.append(computecost(x,y,w,b))     
            # Print cost every at intervals 10 times or as many iterations if < 10
            if i% math.ceil(iterations/10)==0:
             print(f"COST :::{j_history[-1]} , Parameters:: w :{p_history[i][0]} ,b(bias):{p_history[i][1]} ,,Gradients::dj_dw{dj_dw}, dj_db:{dj_db}")    
    return w,b,j_history,p_history
```

```python
w_init=0
b_init=0
iterations = 10000
alpha = 1.0e-2
w_final,b_final,j_history,p_history=gradientdescent(x_train,y_train,w_init,b_init,alpha,iterations,computecost,computegradient)
print(f"\n\n The most accurate parameters w and b found are {w} , {b} respectively")
```
