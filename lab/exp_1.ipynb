```python
dataset["age"]=dataset["age"].astype(int)
```

```python
dataset["alive"]
```

```python
dataset.replace({'yes':1,'no':0,True:1,False:0},inplace=True)
```

```python
dataset.drop(columns=["who","alone","adult_male"],inplace=True)
dataset.head(30)
```

```python
dataset.drop(columns=["embarked"],inplace=True)
```

```python
dataset.head(25)
```

```python
dataset.rename(columns={'sibsp':'siblings_spouse','parch':'parents_children'},inplace=True)
```

```python
dataset.head(50)
```

```python
dataset=pd.get_dummies(dataset,columns=["embark_town"],drop_first=True)
```

```python
dataset.head(15)
```

```python
dataset["age"]=(dataset["age"]-dataset["age"].min())/(dataset["age"].max()-dataset["age"].min())
```

```python
dataset.head(10)
```

```python
dataset.drop(columns=["class"],inplace=True)
```

```python
dataset.head(10)
```

```python
dataset["fare"]=(dataset["fare"]-dataset["fare"].min())/(dataset["fare"].max()-dataset["fare"].min())
```

```python
dataset.head(10)
```

```python
dataset['sex']=dataset['sex'].map({"female":0,"male":1})
```

```python
dataset=pd.get_dummies(dataset,columns=["deck"],drop_first=True)
```

```python
dataset.drop(columns=["alive"],inplace=True)
```

```python
dataset.head(15)
```

```python
dataset = pd.get_dummies(dataset, columns=["pclass"], drop_first=True)
```

```python
dataset["family_size"]=dataset["siblings_spouse"]+dataset["parents_children"]
```

```python
dataset.head(20)
```

```python
dataset.drop(columns=["siblings_spouse","parents_children"],inplace=True)
```

```python
dataset.head(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=dataset.drop(columns=["survived"])
y=dataset["survived"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=3)
```

```python
X_train
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy: ",accuracy_score(y_test,y_pred))
```
